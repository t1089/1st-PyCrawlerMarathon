### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup
import time
 
# 格式化成2016-03-20 11:45:39形式
#print (time.strftime("%Y/%m/%d %H:%M:%S", time.localtime()) )
ttdoday=time.strftime("%Y/%m/%d %H:%M:%S", time.localtime())
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
aa=soup.find(class_="part_list_2").find_all('h3')[0]
aList=[]
aDict={}
bDict={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if(ttdoday[:10]==d.find(class_="date").text[:10]):
        #print(d.find(class_="date").text, d.find(class_='tag').text,d.find('a').text)
        temp=d.find(class_='tag').text
        if(temp not in aList):
            aList.append(temp)
            aDict[temp]=1
            bDict[temp]=[d.find(class_="date").text+' '+d.find('a').text]
        else:
            aDict[temp]+=1
            t2=bDict[temp]
            t2.append(d.find(class_="date").text+' '+d.find('a').text)
            bDict[temp]=t2
#print(aList)
#print(aDict)
#print(bDict)
for key,value in bDict.items():
    print('類別[',key,'] , 共 ',aDict[key],'則')
    for i in value:
        print(i)
```

### ③ 哪一個類別的文章最多

```python
import requests
from bs4 import BeautifulSoup
import time

ttdoday=time.strftime("%Y/%m/%d %H:%M:%S", time.localtime())
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
aa=soup.find(class_="part_list_2").find_all('h3')[0]
aDict={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if(ttdoday[:10]==d.find(class_="date").text[:10]):
        #print(d.find(class_="date").text, d.find(class_='tag').text,d.find('a').text)
        temp=d.find(class_='tag').text
        if(temp not in aDict):
            aDict[temp]=1
        else:
            aDict[temp]+=1
max=1
maxKey=''
for key,value in aDict.items():
    if value>max:
        max=value
        maxKey=key
#print(aDict)
print(maxKey,'共',max,'則')
```
