## requests + BeautifulSoup


```python
# requests + beatifulsoup 實作的方式，改寫成 grab + pyquery
# 取得檔案的類型不一樣，搜尋標籤的方式也不同，執行時間不一同
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'

soup = BeautifulSoup(r.text, "html5lib")
print(soup.find('title').text)
print(type(soup),type(soup.find('title')))

for i in soup.findAll('link'):
    print(i)
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq
g = Grab()
resp = g.go('https://www.dcard.tw/f')

doc = pq(resp.body)
title = doc('title')
print(type(resp.body))
print(type(doc),type(title))
print(title.text())

for i in doc('link').items():
    print(i)
```
