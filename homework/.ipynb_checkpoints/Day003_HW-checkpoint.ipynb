### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
# 讀檔案
fh = open("./data/64_72hr_CH.xml", "r",encoding='utf')
xml = fh.read()
fh.close()

print('File I/O type=',type(xml))
# 解析檔案內容
import xmltodict
d = dict(xmltodict.parse(xml))
print('xmltodict type=',type(d))
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

#print(d['cwbopendata']['dataset']['locations']['location'][0])
location=d['cwbopendata']['dataset']['locations']['location']
print('高雄市共有',len(location),'地區有溫度。')
localName=[]
for i in location:
    localName.append(i['locationName'])
print(localName)
    
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
localNameThing=[]

for i in location:
    localNameThing.append([i['locationName'],
                           i['weatherElement'][0]['time'][0]['dataTime'],
                           i['weatherElement'][0]['time'][0]['elementValue']['value']  ])
for i in localNameThing:
    print(i)
    
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

firstTemp=[]
first=location[0]['weatherElement'][0]['time']
print(location[0]['locationName'])
for i in first:
    firstTemp.append([i['dataTime'],i['elementValue']['value']  ])
print(firstTemp)
```
